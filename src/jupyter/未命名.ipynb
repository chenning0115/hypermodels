```python
import os, sys, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path_prefix = '../res_1024/'
ll = os.listdir(path_prefix)
loa, laa, lkappa, ldata_sign, lsample_num, ln = [], [], [], [], [], []
print(len(ll))
j,k,z = 0,0,0
for line in ll:
    if 'Indian_test' in line:
        items = line.split('_')
        data_sign, sample_num, n = items[2], int(items[0]), int(items[1]) 
        j += 1
    elif 'pavia_cross' in line:
        items = line.split('_')
        data_sign, sample_num, n = items[2], int(items[0]), int(items[1]) 
        k += 1
    elif 'salinas_cross' in line:
        items = line.split('_')
        data_sign, sample_num, n = items[2], int(items[0]), int(items[1]) 
        z += 1
    path = '%s/%s' %(path_prefix, line)
    print(j,k,z)
    with open(path, 'r') as fin:
        ss = json.loads(fin.read())
        oa, aa, kappa = ss['eval']['oa'], ss['eval']['aa'],ss['eval']['kappa']
        for t in ['oa', 'aa', 'kappa', 'data_sign', 'sample_num', 'n']:
            ss = 'l%s.append(%s)' % (t, t)
            eval(ss)
df = pd.DataFrame({
    'data_sign':ldata_sign,
    'sample_num':lsample_num,
    'n':ln,
    'oa': loa,
    'aa':laa,
    'kappa':lkappa,
})
```

```python
df = df.sort_values(['data_sign', 'sample_num', 'n'])
```
