```python
import os, sys, json, string
import numpy as np
import pandas as pd
import copy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_dir(path_prefix, parse_path_fun):
    ll = os.listdir(path_prefix)
    loa, laa, lkappa, ldata_sign, lmethod, lsample_num, ln = [], [], [], [], [], [], []
    each_class = [[] for i in range(30)]

    for line in ll:
        path = '%s/%s' %(path_prefix, line)
        data_sign, sample_num, n, method = parse_path_fun(line)
        with open(path, 'r') as fin:
            ss = json.loads(fin.read())
            if 'eval' in ss:
                ss = ss['eval']
            oa, aa, kappa = float(ss['oa']), float(ss['aa']), float(ss['kappa'])
            if oa <1: oa *= 100
            if aa <1: aa *= 100
            if kappa<1: kappa *= 100
            for t in ['oa', 'aa', 'kappa', 'data_sign', 'sample_num', 'n', 'method']:
                qq = 'l%s.append(%s)' % (t, t)
                eval(qq)
            each = ss['each_acc'].strip('[').strip(']').split(' ')
            temp = []
            for s in each:
                if s == "":
                    continue
                temp.append(s.strip('\n'))
            each = temp
            for i, q in enumerate(each):
                each_class[i].append(float(q))
            for j in range(i+1, len(each_class)):
                each_class[j].append(-1)

    dd = {
        'method': lmethod,
        'data_sign':ldata_sign,
        'sample_num':lsample_num,
        'n':ln,
        'oa': loa,
        'aa':laa,
        'kappa':lkappa,
    }
    for i, each in enumerate(each_class):
        dd['%s' % i] = each
    df_temp = pd.DataFrame(dd)
    return df_temp
```

```python
def parse_path_ours(line):
    j,k,z = 0,0,0
    method = "ours"
    try:
        int(line.split('_')[0])
        if 'Indian_test' in line:
            items = line.split('_')
            data_sign, sample_num, n = items[2], int(items[0]), int(items[1]) 
            j += 1
        elif 'pavia_cross' in line:
            items = line.split('_')
            data_sign, sample_num, n = items[2], int(items[0]), int(items[1]) 
            k += 1
        elif 'salinas_cross' in line:
            items = line.split('_')
            data_sign, sample_num, n = items[2], int(items[0]), int(items[1]) 
            z += 1
    except Exception as e:
        items = line.split('_')
        data_sign,sample_num, n = items[0], int(items[4]), int(items[5])
    return str.lower(data_sign), sample_num, n, method


df_ours = read_dir('../result/res_1024/', parse_path_ours)
```

```python
def parse_fun_compare(line):
    items = line.split('_')
    if 'ss1d' in line or 'hit' in line or 'lsga' in line:
        data_sign, method, sample_num, n = items[0], items[3], int(float(items[1])), int(items[2].strip('.json'))
        if data_sign.strip() == 'Honghu':
            data_sign = 'wh'
    else:
        data_sign, method, sample_num, n = items[0], items[1], int(float(items[2])), int(items[3].strip('.json'))
    data_sign = str.lower(data_sign)
    return data_sign, sample_num, n, method
df_compative = read_dir('../result/res_compative/', parse_fun_compare)
```

```python
def parse_path_wo(line):
    items = line.split('_')
    data_sign, sample_num, n = items[0], int(items[-3]), int(items[-2])
    data_sign = str.lower(data_sign)
    return data_sign, sample_num, n

def parse_path_wo_all(line):
    data_sign, sample_num, n = parse_path_wo(line)
    return data_sign, sample_num, n, 'wo_all'
    
def parse_path_wo_center(line):
    data_sign, sample_num, n = parse_path_wo(line)
    return data_sign, sample_num, n, 'wo_center'

def parse_path_wo_rotate(line):
    data_sign, sample_num, n = parse_path_wo(line)
    return data_sign, sample_num, n, 'wo_rotate'

df_wo_all = read_dir(path_prefix = '../result/res_centerformer_wo_rotate_and_center/', parse_path_fun=parse_path_wo_all)
df_wo_center = read_dir(path_prefix='../result/res_centerformer_wo_center/res_test_centerformer_wo_center/', parse_path_fun=parse_path_wo_center)
df_wo_rotate = read_dir(path_prefix='../result/res_centerformer_wo_rotate_dropout/', parse_path_fun=parse_path_wo_rotate)
```

```python
df = pd.concat([df_ours, df_compative, df_wo_all, df_wo_center, df_wo_rotate])
df_group = df.groupby(['data_sign', 'sample_num', 'method'])
df_avg = df_group.mean()
df_std = df_group.std()
df.to_csv('../data.csv')
```

```python
df_avg = df_avg.reset_index()
```

```python
def plot_one(sign, methods, metric='oa'):
    plt.figure(figsize=(12,6))
    for method in methods:
        df_temp = df_avg[(df_avg['data_sign']==sign)&(df_avg['method']==method)]
        df_temp = df_temp.sort_values(['sample_num'])
        sample_num = df_temp['sample_num']
        oa = df_temp[metric]
        plt.plot(sample_num, oa, label=method)
    plt.legend()
    plt.show()
    
def plot_wo(sign, metric='oa'):
    plt.figure(figsize=(8,4))
    methods = ['ours','wo_all', 'wo_center', 'wo_rotate']
    loa = []
    for method in methods:
        df_temp = df_avg[(df_avg['data_sign']==sign)&(df_avg['method']==method) &(df_avg['sample_num']==10)]
        oa = df_temp[metric]
        loa.append(oa.values[0])
    print(loa)
    plt.bar(methods, loa)
    plt.ylim(85, 100)
    plt.legend()
    plt.show()
    
    
sign = 'wh'
methods = ['SSRN.json', 'casst.json', 'ours', 'ssftt.json', 'sf', 'hit.json', 'lsga.json', 'ss1d.json']
metric = 'oa'
plot_one(sign, methods, metric)

plot_wo(sign, metric)
```
