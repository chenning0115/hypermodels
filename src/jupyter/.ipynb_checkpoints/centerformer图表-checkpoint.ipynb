```python
import os, sys, json, string
import numpy as np
import pandas as pd
import copy
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
df = pd.read_csv('../data.csv')
```

```python
df_group = df.groupby(['data_sign', 'sample_num', 'method'])
df_avg = df_group.mean()
df_std = df_group.std()
df_avg = df_avg.reset_index()
df_std = df_std.reset_index()
```

```python
sample_num = 10
data_sign = 'wh'
class_num = 22

df_avg_temp = df_avg[df_avg['sample_num']==sample_num]
df_std_temp = df_std[df_std['sample_num']==sample_num]
df_avg_temp = df_avg_temp[df_avg_temp['data_sign']==data_sign]
df_std_temp = df_std_temp[df_std_temp['data_sign']==data_sign]
method_dict_revert ={
 'SSRN.json': 'SSRN',
 'casst.json': 'CASST',
 'hit.json':'Hit',
 'lsga.json': 'LSGA',
 'ours': 'Ours',
 'sf': 'SpectralFormer',
 'ss1d.json': 'SS1DSwin',
 'ssftt.json':'SSFTT',
 'wo_all':'wo_all',
 'wo_center': 'wo_center',
 'wo_rotate': 'wo_rotate'}
method_dict = {}
for k,v in method_dict_revert.items():
    method_dict[v] = k

methods = ['SSRN', 'SpectralFormer', 'Hit', 'SS1DSwin', 'SSFTT', 'CASST', 'LSGA', 'Ours']
titles = ['Class'] + methods
label_names=[str(i) for i in range(1,class_num+1)]
```

```python
tex_tbl='\\begin{tabular}{%s} '%('|'.join(['c']*len(titles)))
tex_tbl+='\\toprule '
tex_tbl+='& '.join(titles)+'\\\\ '
tex_tbl+=' \midrule \midrule '

for i in range(class_num):
    line=[]
    line.append(label_names[i])
    data=[]
    max_val = -1
    for method in methods:
        d0 = df_avg_temp[df_avg_temp['method']==method_dict[method]][str(i)].values[0]
        d1 = df_std_temp[df_std_temp['method']==method_dict[method]][str(i)].values[0]
        item = [float(d0),float(d1)]
        data.append(item)
        if item[0] > max_val:
            max_val = item[0]
    for val_avg, val_std in data:
        if val_avg == max_val:
            line.append('\\boldsymbol{$%.2f \\pm %.2f$}' %(val_avg, val_std))
        else:
            line.append('$%.2f \\pm %.2f$' %(val_avg, val_std))
    tex_tbl+='& '.join(line)+'\\\\ '
    
tex_tbl+=' \midrule \midrule '
#开始OA等
for label,nominal in zip(['oa','aa','kappa'],['OA(\%)','AA(\%)','$\kappa \\times 100$']):
    line=[]
    line.append(nominal)
    data=[]
    max_val = -1
    for method in methods:
        d0 = df_avg_temp[df_avg_temp['method']==method_dict[method]][label].values[0]
        d1 = df_std_temp[df_std_temp['method']==method_dict[method]][label].values[0]
        item = [float(d0),float(d1)]
        data.append(item)
        if item[0] > max_val:
            max_val = item[0]
    for val_avg, val_std in data:
        if val_avg == max_val:
            line.append('\\boldsymbol{$%.2f \\pm %.2f$}' %(val_avg, val_std))
        else:
            line.append('$%.2f \\pm %.2f$' %(val_avg, val_std))
    tex_tbl+='& '.join(line)+'\\\\ '
    

tex_tbl+=' \\bottomrule '
tex_tbl+='\end{tabular} '
print(tex_tbl+'\n')
```

```python
def plot_one_method(method, color):
    sample_num = [5,10,15,20,25,30]
    df_avg_temp = df_avg[(df_avg['data_sign']==data_sign) & (df_avg['method']==method_dict[method])]
    df_std_temp = df_std[(df_std['data_sign']==data_sign) & (df_std['method']==method_dict[method])]
    vals = []
    stds = []
    for num in sample_num:
        vals.append(df_avg_temp[df_avg_temp['sample_num']==num]['oa'].values[0])
        stds.append(df_std_temp[df_std_temp['sample_num']==num]['oa'].values[0])
    ls = '-.'
#     plt.errorbar(sample_num, vals, yerr=stds, marker='o', markersize=3, linestyle=ls, label=method)
    print(vals)
    plt.plot(sample_num, vals, marker='^', markersize=6, linestyle=ls, label=method, color=color)

data_sign = 'wh'
show_methods = ['SSRN', 'SpectralFormer', 'Hit', 'SS1DSwin', 'SSFTT', 'CASST', 'LSGA', 'Ours']

plt.figure(figsize=(7,7))
cmap = cm.get_cmap('tab10')
for i, method in enumerate(show_methods):
    plot_one_method(method, cmap(len(show_methods)-i))
legend = plt.legend(loc='lower right', fontsize='small')
plt.show()
```
